```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(data = df , x = "class" , y = "age")
```

```python
sns.barplot(data = df , x = "class" , y = "age" , estimator= "sum")
```

```python
sns.barplot(data = df , x = "class" , y = "age" , hue = "sex")
```

```python
sns.boxplot(data = df , x = "class" , y = "age" )
```

```python
sns.boxplot(data = df , x = "class" , y = "age" , hue = "survived" )
```

```python
sns.violinplot(data = df , x = "class" , y = "age")
```

```python
sns.distplot(df[df["class"] == "Second"]["age"])
```
