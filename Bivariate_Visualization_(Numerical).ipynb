```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.scatterplot(data = df , x = "age" , y = "fare" , hue = "survived")
```

```python
df2 = sns.load_dataset("tips")
df2.head()
```

```python
sns.scatterplot(data = df2 , x = "total_bill" , y = "tip")
```

```python
sns.scatterplot(data = df2 , x = "total_bill" , y = "tip" , hue = "sex")
```

```python
sns.scatterplot(data = df2 , x = "total_bill" , y = "tip" , hue = "sex" , style = "smoker")
```

```python
sns.jointplot(data = df2 , x = "total_bill" , y = "tip")
```

```python
sns.jointplot(data = df2 , x = "total_bill" , y = "tip" ,  hue = "sex")
```

```python
sns.jointplot(data = df2 , x = "total_bill" , y = "tip" , kind = "reg")
```
